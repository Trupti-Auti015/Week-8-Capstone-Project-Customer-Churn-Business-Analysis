 DAY 5- DATA CLEANING & PREPROCESSING

```python


# Import Library
import pandas as pd

# Load Dataset
df = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week8-capstone-business-analysis/data/customer_churn.csv")

print("="*50)
print("DATA CLEANING")
print("="*50)

# Missing Values Before Cleaning
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# Remove Duplicate Records
df = df.drop_duplicates()

# Fill Missing Values
for column in df.columns:
    if df[column].dtype == "object":
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

# Missing Values After Cleaning
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Dataset Shape
print("\nDataset Shape After Cleaning:")
print(df.shape)

# Data Types
print("\nData Types:")
print(df.dtypes)

# Save Cleaned Dataset
df.to_csv(
    "C:/Terraform/TRUPTI/The  developer arena/week8-capstone-business-analysis/data/cleaned_data.csv",
    index=False
)

print("\nCleaned dataset saved successfully!")
```
